```r
# Load ggplot2
library(ggplot2)
```

```r
# Reading the dataset
data(iris)
head(iris)
```

```r
# Plotting the iris dataset
IrisPlot <- ggplot(iris, aes(Sepal.Length, Petal.Length, colour = Species)) + geom_point()
IrisPlot
```

```r
# Modifying the title, x label and y label
print(IrisPlot + labs(y = "Petal length (cm)", x = "Sepal length (cm)")+ggtitle("Petal and sepal length of iris"))
```

#### Highlight and tick marks

```r
# add text
IrisPlot + annotate("text", x = 6, y = 5, label = "Bodderline between the species versicolor and verginica")
```

```r
# Highlight an area
IrisPlot + annotate("rect", xmin = 5, xmax = 7, ymin = 4, ymax = 6, alpha = 0.5)
```

```r
# Segment
IrisPlot + annotate("segment", x = 5, xend = 7, y = 4.125, yend = 5, colour = "black")
```

```r
# Changing attributes for legends
# Remove Legend
IrisPlot + theme(legend.position = "none")
```

```r
# Hide the legend title
IrisPlot + theme(legend.title = element_blank())
```

```r
# Change the legend position
IrisPlot + theme(legend.position = "top")
```

```r
# Adding a regression line
ggplot(iris, aes(Sepal.Length, Petal.Length, colour = Species)) + geom_point() + geom_smooth(method = lm)
```
